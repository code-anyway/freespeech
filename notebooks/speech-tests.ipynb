```python
from freespeech.lib import speech


SSML = """
<speak version="1.0" xmlns="http://www.w3.org/2001/10/synthesis" xml:lang="en-US">
    <voice name="uk-UA-OstapNeural">
        <!-- stress in the word "січними" [sichnymy] should be set on the second vowel, not on the first vowel. -->
        січними
        <!-- stress in the word "площина" should be set on the last vowel, and not on the second vowel. -->
        площина
        <!--  stress in the word "кути" should be made on the last vowel, not on the first vowel. -->
        кути
        <!-- stress in the word "зубочистки" should be made on the third vowel, and not on the second vowel. -->
        зубочистки
        <phoneme alphabet="ipa" ph="sitʃˈnɪmɪ">січними</phoneme>
        <break time="1s" />
        <phoneme alphabet="ipa" ph="ploˈʃtʃɪna">площина</phoneme>
        <break time="1s" />
        <phoneme alphabet="ipa" ph="kuˈtɪ">кути</phoneme>
        <break time="1s" />
        <phoneme alphabet="ipa" ph="zuboˈtʃɪstkɪ">зубочистки</phoneme>
    </voice>
</speak>
"""


data = speech._azure_api_call(SSML)


with open("//tmp/uk-UA-OstapNeural.wav", "wb") as fd:
    fd.write(data)
```
