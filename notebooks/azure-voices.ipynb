```python
from freespeech.lib import speech
from freespeech.types import Voice
import IPython.display as ipd


TEXT = """And you really will have to make it through that violent, metaphysical, symbolic storm.""".replace("\n", "")

file, voice = await speech.synthesize_text(TEXT, duration_ms=None, voice=Voice(character="Alonzo", speech_rate=1.0), lang="en-US", output_dir="/tmp/freespeech")
ipd.Audio(file)
```

```python
file_05, voice = await speech.synthesize_text(TEXT, duration_ms=None, voice=Voice(character="Alonzo", speech_rate=0.5), lang="en-US", output_dir="/tmp/freespeech")
ipd.Audio(file_05)
```
