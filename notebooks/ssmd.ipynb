```python
from freespeech.lib import youtube


res = youtube.download("https://www.youtube.com/watch?v=6ZVccPeStgc", "/tmp/", audio_filename="test.wav", video_filename="test.mp4")
```

```python
from freespeech.lib import speech, text, gdocs
from freespeech.types import Transcript


events = []
acc = None

for event in res[3]["en-US"]:
    assert len(event.chunks) == 1
    if acc == None:
        acc = event
    else:
        acc = speech.concat_events(acc, event, break_sentence=False)

    if text.is_sentence(acc.chunks[0]):
        events += [acc]
        acc = None

if acc:
    events += [acc]


gdocs.create(Transcript(events=events, lang="uk-UA"), format="SSMD-LITE")
        
```
